```python
import pandas as pd

df = pd.read_csv('world-GDP.csv')

df.head(3)
```

```python
df_clean = df.drop(columns=['Series Name', 'Series Code', 'Country Code', '2020 [YR2020]'])
df_clean = df_clean.rename(columns={"1990 [YR1990]": "31-12-1990", "2000 [YR2000]": "31-12-2000",
                                    "2011 [YR2011]": "31-12-2011","1990 [YR1990]": "31-12-1990",
                                    "2012 [YR2012]": "31-12-2012","2013 [YR2013]": "31-12-2013",
                                    "2014 [YR2014]": "31-12-2014","2015 [YR2015]": "31-12-2015",
                                    "2016 [YR2016]": "31-12-2016","2017 [YR2017]": "31-12-2017",
                                    "2018 [YR2018]": "31-12-2018","2019 [YR2019]": "31-12-2019"})
df_clean.head(3)
```

```python
world_GDP = pd.melt(df_clean, 'Country Name', var_name='Date', value_name='GDPperCapGrowth%')
print(world_GDP.head())
```

```python
spain_GDP = world_GDP[world_GDP['Country Name'] == 'Spain']
spain_GDP
```

```python
mexico_GDP = world_GDP[world_GDP['Country Name'] == 'Mexico']
mexico_GDP
```
